```python
from urllib import request
from bs4 import BeautifulSoup

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20211103"
target = request.urlopen(url)

soup = BeautifulSoup(target)

list_title = []
list_point = []

for line in soup.findAll("tr"):
    title = line.find("div", class_ = "tit5")
    
    if title:
        list_title.append(title.get_text().strip("\n"))
    point = line.find("td", class_ = "point")
    if point:
        list_point.append(point.get_text().strip("\n"))

for i, (title, point) in enumerate(zip(list_title, list_point)): # zip은 2개만 가능
    print("{}위 : {} - {}".format(i + 1, title, point))
```

```python
# api test
# naver papago
# request_url = "http://openapi.naver.com/v1/papago/n2mt"
def papago(id_, pw, text, fromlang, tolang):
    import urllib

    encText = urllib.parse.quote(text)
    client_id = id_ # 개발자센터에서 발급받은 Client ID 값
    client_secret = pw # 개발자센터에서 발급받은 Client Secret 값
    data = "source=" + fromlang + "&target=" + tolang + "&text=" + encText
    url = "https://openapi.naver.com/v1/papago/n2mt"
    target = urllib.request.Request(url)
    target.add_header("X-Naver-Client-Id", client_id)
    target.add_header("X-Naver-Client-Secret", client_secret)
    response = urllib.request.urlopen(target, data = data.encode("utf-8"))
    rescode = response.getcode()
    if(rescode == 200):
        response_body = response.read()
        print(response_body.decode('utf-8'))
    else:
        print("Error Code:" + rescode)
```
