### Keras Linear

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
```

```python
W = tf.Variable(tf.random.normal([1]), name = "weight") # 기울기(가중치)
b = tf.Variable(tf.random.normal([1]), name = "bias") # 졀편
hypothesis = x * W + b # f(x) = ax + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y)) # MSE(평균 제곱인데 루트를 씌움)
sgd = tf.keras.optimizers.SGD(learning_rate = 0.01)
```

```python
model = tf.keras.models.Sequential() # 모델 객체 생성
model.add(tf.keras.layers.Dense(1, input_dim = 1)) # 레이어 추가
```

```python
# 모델 생성
model.compile(loss = 'mean_squared_error', optimizer = sgd)
# 모델 학습
model.fit(x, y, epochs=100)
```

```python
model.predict(np.array([6])) # 학습한 내용을 가지고 예측
```

### Logistics - sigmoid
- 1 / 1 + e^-1(ax + b) < x로 편미분

```python
data = [[2, 0], [4, 0], [6, 0], [8, 1], [10, 1], [12, 1], [14, 1]]
x_data = [i[0] for i in data]
y_data = [i[1] for i in data]
```

```python
# 그래프
plt.scatter(x_data, y_data)
plt.xlim(0, 15)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 기울기, 절편, 학습률을 초기화
a = 0; b = 0; lr = 0.05; epochs = 2000;

# 시그모이드 함수 정의
def sigmoid(x):
    return 1 / (1 + np.e ** (-x))
```

```python
for i in range(epochs + 1):
    for x_data, y_data in data:
        # 오차 함수를 a로 편미분
        a_diff = x_data * (sigmoid(a * x_data + b) - y_data)
        b_diff = sigmoid(a * x_data + b) - y_data

        a = a - lr * a_diff # 기울기 update
        b = b - lr * b_diff # 기울기 update
        print("epoch = %.f, 기울기 : %.04f, 절편 : %.04f" %(i, a, b))
```

### 논리 gates

```python
# weight와 bias
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1
```

```python
# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1

# NAND gate
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)
    
# OR gate
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND gate
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w, b3)

# XOR gate : nand와 or를 and 해서 xor가 나옴
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```

```python
# dataset/pima-indians-diabetes.csv를 읽어서
# loss 함수는 'binary_crosentropy', optimizer는 'SGD'
# matrix는 acurrcy, hidden layer는 3개층을, node, epochs는 알아서
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

```python
np.random.seed(2021)
tf.random.set_seed(2021)
```

```python
Data_set = np.loadtxt("./dataset/pima-indians-diabetes.csv", 
                      delimiter = ",")
```

```python
Data_set.shape
```

```python
X = Data_set[:, 0:7]
Y = Data_set[:, 8]
```

```python
model = Sequential()
model.add(Dense(30, input_dim = 7, activation = "relu")) # 30개의 노드
model.add(Dense(30, activation = "relu"))
model.add(Dense(30, activation = "relu"))
model.add(Dense(1, activation = "sigmoid")) 
```

```python
model.compile(loss = "binary_crossentropy", optimizer = "Adam", metrics = ["accuracy"])
model.fit(X, Y, epochs = 300, batch_size = 200)
```
