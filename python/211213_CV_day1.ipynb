```python
img = cv2.imread("./fig/puppy.bmp")
```

```python
cv2.namedWindow("puppy")
cv2.imshow("puppy", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# cv2.imread(filename. flags = None) -> retval
# flags:
# cv2.IMREAD_COLOR: BGR color로 읽기
# cv2.IMREAD_GRAYSCALE: 그레이 color로 읽기
# cv2.IMREAD_UNCHANGED: 파일 속성대로 읽기

# retval: numpy.ndarray로 반환
```

```python
img = cv2.imread("./fig/4560_4215_06.png", cv2.IMREAD_GRAYSCALE)

if img is None:
    print("image is not found")
    sys.exit()

cv2.namedWindow("image", cv2.WINDOW_AUTOSIZE)
# cv2.resizewindow("image", 400, 300)

cv2.imshow("image", img)

while True:
    if cv2.waitKey() == 27:
        break
# cv2.waitKey(3000) # 동영상 일 때 3000

cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
if img is None:
    print("image read failed")
    sys.exit()

# BGR to RGB
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imggray = cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE)

plt.imshow(imgrgb)
# plt.axis("off")
plt.show()
```

```python
plt.figure(figsize = (15, 15))
plt.subplot(131), plt.imshow(img)
plt.subplot(132), plt.imshow(imgrgb)
plt.subplot(133), plt.imshow(imggray, cmap = "gray")

plt.show()
```
