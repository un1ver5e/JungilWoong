```python
img_list = os.listdir("./images/")
img_files = []
for i in img_list:
    img_file = "./images/" + i
    img_files.append(img_file)
    
cv2.namedWindow("scene", cv2.WINDOW_NORMAL)
cv2.setWindowProperty("scene", cv2.WND_PROP_FULLSCREEN,
                     cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_files[idx])
    cv2.imshow("scene", img)
    
    if cv2.waitKey(2000) == 27:
        break
        
    idx += 1
    
    if idx >= len(img_list):
        idx = 0
    
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("./images/cow.png")
img2 = cv2.imread("./images/green.png")
img_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

_, mask = cv2.threshold(img_gray, 240, 255, cv2.THRESH_BINARY_INV) # 마스크 영상 return

h, w = img1.shape[:2]
img2 = cv2.resize(img2, (w, h), cv2.INTER_AREA)

cv2.copyTo(img1, mask, img2) # 합성 됨

cv2.imshow("img2", img2)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread("./fig/puppy.bmp", 1)
sunglass = cv2.imread("./images/imgbin_sunglasses_1.png", cv2.IMREAD_UNCHANGED)
sunglass = cv2.resize(sunglass, (350, 150), cv2.INTER_AREA)

mask = sunglass[:, :, 3]
glass = sunglass[:, :, 0:3]

h, w = mask.shape[:2]

crop = src[120:120 + h, 220:220 + w]

cv2.copyTo(glass, mask, crop)

crop[mask > 0] = (0, 0, 255)

cv2.imshow("src", src)
cv2.imshow("mask", mask)
cv2.imshow("glass", glass)
cv2.imshow("crop", crop)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = np.ones((600, 1200, 3), np.uint8) * 255

# line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
cv2.line(img, (50, 50), (300, 50), (0, 0, 255), 5, cv2.LINE_4) # cv2의 렬행, BGR 주의
cv2.line(img, (300, 50), (150, 300), (0, 0, 255), 5, cv2.LINE_8)
cv2.line(img, (400, 50), (400, 300), (0, 0, 255), 5, cv2.LINE_AA)
# arrowedLine(img, pt1, pt2, color[, thickness[, line_type[, shift[, tipLength]]]])
cv2.arrowedLine(img, (50, 40), (300, 40), (255, 0, 0), 5)
# rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
cv2.rectangle(img, (100, 220), (380, 380), (255, 0, 0), 5) 
# 좌측 상단, 우측 하단, 5 자리에 -1 들어가면 채워줌
cv2.rectangle(img, (50, 350, 350, 150), (0, 255, 0), -1)

cv2.circle(img, (600, 200), 100, (255, 255, 0), 5, cv2.LINE_AA) # LINE_AA 안티 얼라이싱

text = "opencv version, " + cv2.__version__

cv2.putText(img, text, (800, 100), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 1, 
            cv2.LINE_AA)

cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
